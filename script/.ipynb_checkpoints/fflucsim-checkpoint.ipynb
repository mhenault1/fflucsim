```python
assays_info = pd.read_csv('assays_info.csv')
assays_info['log10_monosome_rate'] = assays_info['monosome_rate'].apply(lambda x: f'{np.log10(x):.1f}')
assays_info['log10_revert_rate'] = assays_info['revert_rate'].apply(lambda x: f'{np.log10(x):.1f}')

populations_info = pd.read_csv('populations_info.csv')
populations_info['log10_monosome_rate'] = populations_info['monosome_rate'].apply(lambda x: f'{np.log10(x):.1f}')
populations_info['log10_revert_rate'] = populations_info['revert_rate'].apply(lambda x: f'{np.log10(x):.1f}')
```

```python
FA_Store = {}
FA_Results = []

for a in assays_info['assay'].values:

    fn = f'fa.{a}.pkl'
    fp = f'../data/pickle/{fn}'
    if fn in os.listdir('../data/pickle/'):
        with open(fp, 'rb') as fi:
            FA = pkl.load(fi)
            
            for r in FA.Results:
                r['assay'] = a
                FA_Results.append(pd.Series(r))
            FA_Store[a] = FA

FA_Results = pd.DataFrame(FA_Results).merge(assays_info, on='assay')
```

```python
mutant_count_matrix = []
for a in assays_info['assay']:

    mc = np.flip(np.sort(FA_Store[a].m_monosome))
    mutant_count_matrix.append(mc)
mutant_count_matrix = np.array(mutant_count_matrix)
```

```python
sns.heatmap(mutant_count_matrix, robust=True)
```

```python
for w in (1.0, 0.6):
    
    df = FA_Results.set_index(['monosome_fitness', 'mutant', 'model']).sort_index().loc[w]
    
    fig, axes = plt.subplots(ncols=2, figsize=[10,5])
    
    for ax_idx, mutant in zip((0,1), ('monosome', 'revert')):
        ax = axes[ax_idx]
        if w == 1:
            model = 'LD'
        else:
            model = 'MK'
            
        matrix = df.loc[(mutant, model)].pivot_table(index='log10_monosome_rate', columns='log10_revert_rate', values='mu').map(np.log10)
    
        sns.heatmap(matrix, ax=ax, vmin=-9, vmax=-4, annot=True)
        ax.set_title(f'w={w}\nmutant:{mutant}\nmodel:{model}')
        ax.tick_params(axis='x', rotation=90)
        ax.tick_params(axis='y', rotation=0)
    
    plt.tight_layout()
    plt.show()
    plt.close()
    
```

## Generate metadata sheet of FA params

```python
def ngen_monosome_rate(mr, x=20):
    n_gen = np.int64(np.ceil(np.log2(1/mr*x)))
    return min([n_gen, 25])
```

```python
assays_info = []
populations_info = []

monosome_fitness = np.linspace(0.1, 1, 10)
monosome_fitness = np.concat([monosome_fitness, np.logspace(-3, -1.3, 3)])
monosome_rates = np.logspace(-7, -4, 10)
reversion_rates = np.logspace(-7, -1, 10)
n_reps = 36
assay_idx = 0
pop_idx = 0

for (mf, mr, rr) in itertools.product(monosome_fitness, monosome_rates, reversion_rates):
    assay_id = f'a{assay_idx}'
    assay_idx += 1
    assays_info.append([assay_id, n_reps, mf, mr, rr, ngen_monosome_rate(mr)])
    
    for n in range(n_reps):
        populations_info.append([assay_id, n, mf, mr, rr, ngen_monosome_rate(mr), f'p{pop_idx}'])
        pop_idx += 1

assays_info = pd.DataFrame(assays_info, columns=['assay', 'n_rep', 'monosome_fitness', 'monosome_rate', 'revert_rate', 'n_gen'])
populations_info = pd.DataFrame(populations_info, columns=['assay', 'rep', 'monosome_fitness', 'monosome_rate', 'revert_rate', 'n_gen', 'pop_id'])

#assays_info.to_csv('assays_info.csv', index=False)
#populations_info.to_csv('populations_info.csv', index=False)
```
